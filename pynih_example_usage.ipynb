```python
from pynih import apis

# define search criteria
search_criteria ={
        "agencies":["NIAID"],
        "include_active_projects": True
    }

# illustration of project api usage
include_fields = [ "ApplId","SubprojectId","FiscalYear","OrgName","OrgCity", "OrgState","OrgStateName","DeptType", "ProjectNum","OrgCountry" ]
data = apis.query_project_api(include_fields=include_fields,
                              search_criteria=search_criteria,
                              limit=500)
```

```python
import pandas as pd
pd.DataFrame(data)
```
